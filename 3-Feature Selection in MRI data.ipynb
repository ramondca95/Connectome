## <center> Selección de características importantes usando ANOVA </center>
___

Para filtrar las más importantes usaremos el P-Value

```python
from functions_stats import value_anova

p_value = value_anova(data_and_target, features[4], "p")
print('Region: \t' + str(lista_region_name[0]))
print('Feature: \t' + str(features[4]))
print('P value: \t' + str(p_value))
```

### Con la función *all_values_anova()* obtendremos la tabla con los valores p y F, usando como umbral el P valor.

```python
from functions_stats import all_values_anova

help(all_values_anova)
```

```python
df_reg_feat_anova_05 = all_values_anova(mri_completed, lista_region_name, features, 0.05)
df_reg_feat_anova_05
```

### Con la función *join_feature_region()* uniremos los nombres de la región y característica para crear UNA SOLA CARACTERÍSTICA.

```python
from functions_dataframe import join_feature_region

help(join_feature_region)
```

___
### PRINCIPALES CARACTERÍSTICAS CON P-VALUE < 0.05

```python
df_feat_anova05 = join_feature_region(df_reg_feat_anova_05)
df_feat_anova05
```

```python
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (15,9)
plt.rcParams.update({'font.size': 22})

df_feat_anova05.plot(x = 'Feature_Region', y = 'F-value', kind = 'bar')
plt.title('Principales características usando ANOVA y P-Value < 0.05')
plt.xlabel('Características')
plt.ylabel('F-Value score')
plt.show()
plt.close()
```

___
### PRINCIPALES CARACTERÍSTICAS CON P-VALUE < 0.01

```python
# | Reg | Feat |
df_reg_feat_anova_01 = all_values_anova(mri_completed, lista_region_name, features, 0.01)

# | Reg + Feat |
df_feat_anova01 = join_feature_region(df_reg_feat_anova_01)

# Plotear gráfica con F-value. Ordenar de manera ascendente.
df_feat_anova01.sort_values('F-value', ascending=True).plot(x = 'Feature_Region', y = 'F-value', kind = 'barh')

plt.title('Principales características usando ANOVA y P-Value < 0.01')
plt.xlabel('Características')
plt.ylabel('F-Value score')
plt.show()
plt.close()
```
