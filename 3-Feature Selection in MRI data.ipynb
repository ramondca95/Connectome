```python
import seaborn as sns

def sns_boxplot_df(df):
    """
    Función para plotear bloxplot de seaborn pero con la configuración deseada.
    :param df: Dataframe _boxplot
    :return:
    """
    for i in range(0,len(list(df)) -1):
        sns.boxplot(x = df.sort_values('Label', ascending=True)['Label'], y = df.sort_values('Label', ascending=True).iloc[:,i], palette = ['b','b','r','r'])
        sns.swarmplot(x = df.sort_values('Label', ascending=True)['Label'], y = df.sort_values('Label', ascending=True).iloc[:,i], data=df, color=".05")
        plt.title('Boxplot OS: Original Set vs VS: Validation Set')
        plt.xlabel('Conjuntos de datos')
        plt.show()
        plt.close()
help(sns_boxplot_df)
```

```python
sns_boxplot_df(df_feat_anova01_boxplot)
```
